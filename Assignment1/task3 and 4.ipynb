**Task 4: Proving Symmetry and Positive Semidefiniteness of Covariance Matrix**

To demonstrate that the covariance matrix (C) for any random variable (X in R^p) possesses the properties of symmetry and positive semidefiniteness, i.e., (C = C^T) and (x^T Cx >= 0) for any covariance matrix (C in R^{p x p}) and any vector (x in R^p), we can express this mathematically:

1. **Symmetry (C = C^T):**
   
   Because for any covariance matrix C, it holds that C = C^T.

2. **Positive Semidefiniteness (x^T Cx >= 0):**
   
   Because for any covariance matrix C and any vector x, it is true that x^T Cx >= 0.

**Symmetry Proof:**


# Proving C^T = C:

Given the covariance matrix C = E((X-μ)(X-μ)^T), we want to show that C^T = C.

1. Expand the covariance matrix:
   \[ C = E((X-μ)(X-μ)^T) \]
   \[ = E((X-μ)(X^T-μ^T)) \]
   \[ = E(XX^T-Xμ^T-μX^T+μμ^T) \]
   \[ = E(XX^T)-E(Xμ^T)-E(μX^T)+E(μμ^T) \]
   \[ = E(XX^T)-μμ^T \]

2. Transpose the covariance matrix:
   \[ C^T = E(XX^T)^T-μμ^T \]
   \[ = E((X^T)^T(X)^T)-μ^Tμ^T \]
   \[ = E(XX^T)-μμ^T \]

Therefore, C^T = C.

# Proving x^T C x >= 0:

Now, let's prove that x^T C x >= 0 for any vector x.

1. Start with the definition of C:
   \[ C = E((X-μ)(X-μ)^T) \]

2. Consider x^T C x:
   \[ x^T C x = x^T E((X-μ)(X-μ)^T) x \]
   \[ = E(x^T(X-μ)(X-μ)^T x) \]
   \[ = E((x^T(X-μ))(x^T(X-μ))^T) \]
   \[ = E(Y Y^T) \quad \text{where } Y = x^T(X-μ) \]

3. Since Y is a vector, Y^T Y is the squared magnitude of Y, which is non-negative:
   \[ x^T C x = E(Y^2) \geq 0 \]

Therefore, x^T C x >= 0.
